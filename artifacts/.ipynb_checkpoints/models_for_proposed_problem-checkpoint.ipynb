```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
# 1. 加载数据
# 建议在本地使用 pd.read_parquet('chicago_crime_2015_2024_enriched.parquet')
df = pd.read_parquet('C:\\Users\\LENOVO\\Downloads\\chicago_crime_2015_2024_enriched.parquet')

# 2. 数据转换：按周和社区区域进行聚合
df['Date'] = pd.to_datetime(df['Date'])
df['Week'] = df['Date'].dt.to_period('W').apply(lambda r: r.start_time)

# 统计每周、每个社区的犯罪总数
weekly_crime = df.groupby(['Week', 'Community Area']).size().reset_index(name='crime_count')

# 3. 定义“高风险”标签
# 我们将犯罪数超过全市该周 75% 分位数的区域定义为 High-Risk (1)
weekly_crime['is_high_risk'] = weekly_crime.groupby('Week')['crime_count'].transform(
    lambda x: (x >= x.quantile(0.75)).astype(int)
)

# 4. 特征工程 (Feature Engineering)
# 滞后特征：获取过去几周的犯罪数据（模型无法提前看到未来，所以必须用 Lag）
weekly_crime = weekly_crime.sort_values(['Community Area', 'Week'])
for i in [1, 2, 4]:
    weekly_crime[f'lag_crime_{i}'] = weekly_crime.groupby('Community Area')['crime_count'].shift(i)

# 滚动平均特征：过去4周的平均水平
weekly_crime['rolling_mean_4'] = weekly_crime.groupby('Community Area')['crime_count'].transform(
    lambda x: x.shift(1).rolling(window=4).mean()
)

# 时间周期特征
weekly_crime['month'] = weekly_crime['Week'].dt.month
weekly_crime['week_of_year'] = weekly_crime['Week'].dt.isocalendar().week.astype(int)

# 删除包含空值的行（由于 Shift 产生）
weekly_crime = weekly_crime.dropna()

# 5. 划分训练集和测试集 (采用时间划分，模拟真实预测)
# 假设用 2023 年及以前的数据训练，2024 年的数据测试
train = weekly_crime[weekly_crime['Week'] < '2024-01-01']
test = weekly_crime[weekly_crime['Week'] >= '2024-01-01']

features = ['Community Area', 'lag_crime_1', 'lag_crime_2', 'lag_crime_4', 'rolling_mean_4', 'month', 'week_of_year']
target = 'is_high_risk'

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

# 6. 构建并训练 LightGBM
clf = lgb.LGBMClassifier(
    n_estimators=1000,
    learning_rate=0.05,
    num_leaves=31,
    is_unbalance=True,  # 自动调整类别权重
    importance_type='gain'
)

clf.fit(X_train, y_train, eval_set=[(X_test, y_test)], eval_metric='auc', callbacks=[lgb.early_stopping(stopping_rounds=50)])

# 7. 性能测试 (Performance Testing)
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:, 1]
plt.rcParams['font.sans-serif'] = ['SimHei']
print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))

print(f"ROC-AUC Score: {roc_auc_score(y_test, y_prob):.4f}")

# 8. 特征重要性分析
lgb.plot_importance(clf, max_num_features=10, title="The Importance of Features in Predicting High-Risk Areas")
plt.show()
model_filename = 'chicago_crime_model_lightgbm.joblib'
joblib.dump(model_xgb, model_filename)

print(f"模型已成功保存为: {model_filename}")
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
# 解决绘图中文乱码问题 (Windows 环境常用字体)
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

# 1. 加载数据
path = r'C:\Users\LENOVO\Downloads\chicago_crime_2015_2024_enriched.parquet'
df = pd.read_parquet(path)

# 2. 数据转换：按周和社区区域进行聚合
df['Date'] = pd.to_datetime(df['Date'])
df['Week'] = df['Date'].dt.to_period('W').apply(lambda r: r.start_time)
weekly_crime = df.groupby(['Week', 'Community Area']).size().reset_index(name='crime_count')

# 3. 定义“高风险”标签 (与上次一致，取 75% 分位数)
weekly_crime['is_high_risk'] = weekly_crime.groupby('Week')['crime_count'].transform(
    lambda x: (x >= x.quantile(0.75)).astype(int)
)

# 4. 特征工程
weekly_crime = weekly_crime.sort_values(['Community Area', 'Week'])
for i in [1, 2, 4]:
    weekly_crime[f'lag_crime_{i}'] = weekly_crime.groupby('Community Area')['crime_count'].shift(i)

weekly_crime['rolling_mean_4'] = weekly_crime.groupby('Community Area')['crime_count'].transform(
    lambda x: x.shift(1).rolling(window=4).mean()
)

weekly_crime['month'] = weekly_crime['Week'].dt.month
weekly_crime['week_of_year'] = weekly_crime['Week'].dt.isocalendar().week.astype(int)
weekly_crime = weekly_crime.dropna()

# 5. 划分数据集 (时间划分)
train = weekly_crime[weekly_crime['Week'] < '2024-01-01']
test = weekly_crime[weekly_crime['Week'] >= '2024-01-01']

features = ['Community Area', 'lag_crime_1', 'lag_crime_2', 'lag_crime_4', 'rolling_mean_4', 'month', 'week_of_year']
target = 'is_high_risk'

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

# 6. 构建 XGBoost 模型
# 计算正负样本比例，用于处理不平衡数据
ratio = float(np.sum(y_train == 0)) / np.sum(y_train == 1)

model_xgb = xgb.XGBClassifier(
    n_estimators=1000,
    max_depth=6,
    learning_rate=0.05,
    scale_pos_weight=ratio,  # 处理类别不平衡的关键参数
    tree_method='hist',      # 使用直方图算法加速
    early_stopping_rounds=50,
    eval_metric=['auc', 'logloss'],
    random_state=42
)

# 7. 训练模型
model_xgb.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)

# 8. 性能测试
y_pred = model_xgb.predict(X_test)
y_prob = model_xgb.predict_proba(X_test)[:, 1]

print("\n--- XGBoost Classification Report ---")
print(classification_report(y_test, y_pred))
print(f"XGBoost ROC-AUC Score: {roc_auc_score(y_test, y_prob):.4f}")

# 9. 可视化特征重要性
plt.figure(figsize=(10, 6))
xgb.plot_importance(model_xgb, max_num_features=10, importance_type='gain')
plt.title("XGBoost Feature Importance (Gain)")
plt.show()
model_filename = 'chicago_crime_model_xgboost.joblib'
joblib.dump(model_xgb, model_filename)

print(f"模型已成功保存为: {model_filename}")
```
